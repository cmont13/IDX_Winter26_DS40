```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt
```

# Load Data

```python
X_train = pd.read_csv('property_data_train_x.csv')
X_test = pd.read_csv('property_data_test_x.csv')

y_train = pd.read_csv('property_data_train_y.csv')
y_test = pd.read_csv('property_data_test_y.csv')
```

# Linear Regression Model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Evaluation on Test Data



```python
y_pred = lr.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"LINEAR REGRESSION EVALUATION METRICS\n\nR-Squared: {r2:.3f}\nMean Squared Error: {mse:.2f}\nMean Absolute Error: {mae:.2f}\nRoot Mean Squared Error: {rmse:.2f}\n")
```

## Feature Importance

```python
fig, ax = plt.subplots()

feature_importance = feature_importance = pd.DataFrame({'feature': list(X_train.columns), 'importance': np.abs(lr.coef_[0])}).sort_values('importance', ascending=False).head(10)

plt.barh(feature_importance['feature'], feature_importance['importance'])

ax.set_xlabel('Feature Importance')
ax.set_ylabel('Feature')
ax.set_title('Most Importance Features')
ax.invert_yaxis()

plt.show()
```
